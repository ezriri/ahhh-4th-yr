```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from PIL import Image
import matplotlib.patches as patches
from matplotlib.patches import ConnectionPatch
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

import matplotlib.gridspec as gridspec

# bounds for la-palma-nth:
# min lat: 28.8735
# max lat: 28.6785
# min long: -18.0176
# max long: -17.7209

df_1 = pd.read_csv('transect_1_processed.csv',usecols=['Lat','Long']) ## just got lat + long
df_2 = pd.read_csv('transect_2_processed.csv',usecols=['Lat','Long']) ## just got lat + long

# boundaries of the map image:
#BBox = ((df.Long.min(), df.Long.max(), df.Lat.min(),df.Lat.max()))
##      long max | long min | lat min | lat max 
BBox = [-18.0148, -17.7209, 28.4500, 28.8651]
BBoxW = [-17.9289, -17.8734, 28.7579, 28.7929]
BBoxE = [-17.8384, -17.7694, 28.7114, 28.7525]

map = plt.imread('the-palma.png')
## these two have points on
east_locs = plt.imread('east_locs.png')
west_locs = plt.imread('west_locs.png')

# these two dont
east = plt.imread('east-tran.png')
west = plt.imread('west-tran.png')
fig = plt.figure(tight_layout = True, figsize =(9,7))

gs = gridspec.GridSpec(2, 2,width_ratios=[2.5, 1], height_ratios=[1, 3])

west_map_plot = plt.subplot(gs[0])
compass = plt.subplot(gs[1])
main_map_plot = plt.subplot(gs[2])
east_map_plot = plt.subplot(gs[3])



main_map_plot.set_xlim((BBox[0],BBox[1]))
main_map_plot.set_ylim((BBox[2],BBox[3]))
main_map_plot.imshow(map, zorder = 0, extent = BBox)
# make rectangle outlines (using the already known box dimentions)
rec_e = patches.Rectangle((BBoxE[1],BBoxE[2]), BBoxE[0]-BBoxE[1], BBoxE[3]-BBoxE[2], linewidth = 1, edgecolor = 'k', facecolor = 'none')
main_map_plot.add_patch(rec_e)
rec_w = patches.Rectangle((BBoxW[1],BBoxW[2]), BBoxW[0]-BBoxW[1], BBoxW[3]-BBoxW[2], linewidth = 1, edgecolor = 'k', facecolor = 'none')
main_map_plot.add_patch(rec_w)
main_map_plot.set_title('a)', x=0.93, y=0.93)


west_map_plot.tick_params(left = False, right = False , labelleft = False,labelbottom = False, bottom = False)
west_map_plot.set_xlim(BBoxW[0],BBoxW[1])
west_map_plot.set_ylim(BBoxW[2],BBoxW[3])
west_map_plot.imshow(west, zorder = 2, extent = BBoxW)
west_map_plot.scatter(df_2.Long, df_2.Lat,zorder = 2, c = 'k', marker= 'x')
west_map_plot.set_title('c) West', x=0.83, y=0.83)


#east_map_plot.axis('off')
east_map_plot.tick_params(left = False, right = False , labelleft = False,labelbottom = False, bottom = False)
east_map_plot.scatter(df_1.Long, df_1.Lat,zorder = 1, c = 'k', marker= 'x')
east_map_plot.set_xlim(BBoxE[0],BBoxE[1])
east_map_plot.set_ylim(BBoxE[2],BBoxE[3])
east_map_plot.imshow(east, zorder = 0, extent = BBoxE)
east_map_plot.set_title('b) East', x=0.83, y=0.83)


compass.axis('off')

### this bit of stuff adds all the lines connecting main map --> subplots
# letters: w / e = west / east, t / b = top / bottom, l / r = left / right
wtl = (BBoxW[0],BBoxW[3])
wtr = (BBoxW[1],BBoxW[3])
wbl = (BBoxW[0],BBoxW[2])
wbr = (BBoxW[1],BBoxW[2])

etl = (BBoxE[0],BBoxE[3])
etr = (BBoxE[1],BBoxE[3])
ebl = (BBoxE[0],BBoxE[2])
ebr = (BBoxE[1],BBoxE[2])

# basically we can connect from within the main plot to the sub plots
w_1 = ConnectionPatch(xyA=wbl, coordsA=main_map_plot.transData, xyB=wbl, coordsB=west_map_plot.transData)
w_2 = ConnectionPatch(xyA=wtr, coordsA=main_map_plot.transData, xyB=wtr, coordsB=west_map_plot.transData)
w_3 = ConnectionPatch(xyA=wtl, coordsA=main_map_plot.transData, xyB=wtl, coordsB=west_map_plot.transData)
w_4 = ConnectionPatch(xyA=wbr, coordsA=main_map_plot.transData, xyB=wbr, coordsB=west_map_plot.transData)

e_1 = ConnectionPatch(xyA=etr, coordsA=main_map_plot.transData, xyB=etr, coordsB=east_map_plot.transData)
e_2 = ConnectionPatch(xyA=ebl, coordsA=main_map_plot.transData, xyB=ebl, coordsB=east_map_plot.transData)
e_3 = ConnectionPatch(xyA=ebr, coordsA=main_map_plot.transData, xyB=ebr, coordsB=east_map_plot.transData)
e_4 = ConnectionPatch(xyA=etl, coordsA=main_map_plot.transData, xyB=etl, coordsB=east_map_plot.transData)

#### this load of bollocks makes sure the connecting lines don't overlap the pictures 
## set_zorder = the layer we want to put the pic on i.e. 0 is the bottom // 2 is the upper most layer
main_map_plot.set_zorder(0)
west_map_plot.set_zorder(2)
east_map_plot.set_zorder(2)

fig.add_artist(w_1)
wline_2 = fig.add_artist(w_2)
wline_3 = fig.add_artist(w_3)
wline_4 = fig.add_artist(w_4)
wline_2.set_zorder(1)
wline_3.set_zorder(1)
wline_4.set_zorder(1)

fig.add_artist(e_1)
fig.add_artist(e_2)
eline_3 = fig.add_artist(e_3)
eline_4 = fig.add_artist(e_4)
eline_3.set_zorder(1)
eline_4.set_zorder(1)

## need to make an explicit ledgend:
site_marker = [Line2D([0], [0], linewidth=0, marker='x', color='k')]
label = ['Site location']
compass.legend(site_marker, label, loc = 'center')


plt.savefig('map_of_loc.png')
```
